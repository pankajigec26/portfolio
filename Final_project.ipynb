# 1. KNN Algoritham

```python
from sklearn.model_selection import train_test_split
#splitting the data into four variables , 70% of data will be used to train the model and rest 30% is to test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#Getting the shapes of data
print (X_test.shape)
print (X_train.shape)
print (y_test.shape)
print (y_train.shape)
```

```python
#importing KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier


model = KNeighborsClassifier(n_neighbors=10)
model.fit(X_train,y_train)
```

```python
#doing the prediction 
predicted= model.predict(X_test) # 0:Overcast, 2:Mild
print(predicted)
```

```python
#Finding the accuracy score 
score=round(model.score(X_test,y_test)*100,2)
#Score for predicting the survivor 
print (score)
```

```python
#Confusion matrix
from sklearn import metrics
print (metrics.confusion_matrix(predicted,y_test))
```

Creating a function to return precison and recall 

```python
#recall and precision
from sklearn import metrics
def myscores(smat): 
    tp = smat[0][0] 
    fp = smat[0][1] 
    fn = smat[1][0] 
    tn = smat[1][1] 
    return tp/(tp+fp), tp/(tp+fn)

recall=round(myscores(metrics.confusion_matrix(predicted,y_test))[0]*100,2)
precision=round(myscores(metrics.confusion_matrix(predicted,y_test))[1]*100,2)
print ('recall is ==>'+str(recall))
print ('precision is ==>'+str(precision))
```

```python
data_knn=[['KNN',score,precision,recall]]
```
